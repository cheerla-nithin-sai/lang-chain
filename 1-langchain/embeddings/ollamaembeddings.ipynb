```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
# internally you have to download ollama from website into local machine
# ollama supports embeddings models, making it easy to build RAG apps that combine text prompts with with existing documents
```

```python
ollama_embeddings = OllamaEmbeddings(model="gemma:2b")
# by def model uses lemma2 but internally we downloaded gemma:2b model
```

```python
ollama_text = "this is the text given to ollama"
query_result = ollama_embeddings.embed_documents([ollama_text])
query_result
```

```python
len(query_result[0])
# constant 2048 dimensions in ollama gemma:2b
```
