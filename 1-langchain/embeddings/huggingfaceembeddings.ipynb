```python
import os
from dotenv import load_dotenv
# to load all the env variables
load_dotenv()
```

```python
os.environ['HF_TOKEN'] = os.getenv("HF_TOKEN")
```

```python
# setting up the model in hf embeddings
from langchain_huggingface import HuggingFaceEmbeddings
hf_embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# trying embedding on simple text
hf_example_text = "This is the first text that needs to be converted into vectors"
query_result = hf_embeddings.embed_query(hf_example_text)
query_result
```
